```python
import pandas as pd
sfr_comb = pd.read_csv('Singapore_flat_data.csv')
```

```python
# Data Analysis
# Average resale price for each flat type
import matplotlib.pyplot as plt
avg_flty = sfr_comb.groupby('flat_type')['resale_price'].mean()
leg_flty = list(sfr_comb.flat_type.unique())
avg_flty_lis = [x for x in avg_flty]
plt.figure(figsize=(8, 8))
plt.barh(leg_flty, avg_flty_lis, color='yellow')
plt.ylabel('Flat Type')
plt.xlabel('Average resale price (in SGD)')
plt.title('Average resale price of each flat type')
for i, v in enumerate(avg_flty_lis):
    plt.text(v*0.1, i, str(round(v,2)), va='center', color='black')
plt.show()
```

```python
# Average resale price for each flat model
avg_flmd = sfr_comb.groupby('flat_model')['resale_price'].mean()
leg_flmd = list(sfr_comb.flat_model.unique())
avg_flmd_lis = [x for x in avg_flmd]
plt.figure(figsize=(8, 8))
plt.barh(leg_flmd, avg_flmd_lis, color='skyblue')
plt.ylabel('Flat Model')
plt.xlabel('Average resale price (in SGD)')
plt.title('Average resale price of each flat model')
for i, v in enumerate(avg_flmd_lis):
    plt.text(v*0.1, i, str(round(v,2)), va='center', color='black')
plt.show()
```

```python
# Removing unnecessary columns
for x in ['month','block','street_name']:
    del sfr_comb[x]
```

```python
# Encoding of categorical variables
categorical_columns = sfr_comb.select_dtypes(include=['object']).columns.tolist()
from sklearn.preprocessing import OneHotEncoder
one_hot = OneHotEncoder(sparse_output=False)
encoded_labels = one_hot.fit_transform(sfr_comb[categorical_columns])
```

```python
sfr_encoded = pd.DataFrame(encoded_labels,columns=one_hot.get_feature_names_out(categorical_columns))
```

```python
sfr_encoded
```

```python
sfr_comb1 = sfr_comb.drop(categorical_columns,axis=1)
```

```python
sfr_comb_encd = pd.concat([sfr_comb1,sfr_encoded],axis=1)
sfr_comb_encd
```

```python
# Linear Regression - to predict resale price
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,root_mean_squared_error
X = sfr_comb_encd.drop(['resale_price'],axis=1)
Y = sfr_comb_encd['resale_price']
# Normalization
scale = StandardScaler()
X = scale.fit_transform(X.values)
# Splitting the data into train and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
# Linear Regression model
model_lin = LinearRegression()
# Train the model
model_lin.fit(X_train,Y_train)
# Get predictions
Y_pred = model_lin.predict(X_test)
# Accuracy of predictions
rmse_lin = root_mean_squared_error(Y_test,Y_pred)
r2_lin = r2_score(Y_test,Y_pred)
print(f'Root mean squared error: {rmse_lin}')
print(f'R2 score: {r2_lin}')
```

```python
# Decision Tree Regressor - to predict resale price
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score,root_mean_squared_error
X = sfr_comb_encd.drop(['resale_price'],axis=1)
Y = sfr_comb_encd['resale_price']
# Normalization
scale = StandardScaler()
X = scale.fit_transform(X.values)
# Splitting the data into train and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
# defining the model
model_dec = DecisionTreeRegressor()
# Train the model
model_dec.fit(X_train,Y_train)
# Get predictions
Y_pred = model_dec.predict(X_test)
# Accuracy of predictions
rmse_dec = root_mean_squared_error(Y_test,Y_pred)
r2_dec = r2_score(Y_test,Y_pred)
print(f'Root mean squared error: {rmse_dec}')
print(f'R2 score: {r2_dec}')
```

```python
# Random Forest Regressor - to predict resale price
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,root_mean_squared_error
X = sfr_comb_encd.drop(['resale_price'],axis=1)
Y = sfr_comb_encd['resale_price']
# Normalization
scale = StandardScaler()
X = scale.fit_transform(X.values)
# Splitting the data into train and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
# Defining the model
model_rand = RandomForestRegressor()
# Train the model
model_rand.fit(X_train,Y_train)
# Get predictions
Y_pred = model_rand.predict(X_test)
# Accuracy of predictions
rmse_rand = root_mean_squared_error(Y_test,Y_pred)
r2_rand = r2_score(Y_test,Y_pred)
print(f'Root mean squared error: {rmse_rand}')
print(f'R2 score: {r2_rand}')
```

```python
# The root mean squared error is less and R2 score is more for Random Forest Regressor, so we choose this model.
# Predict the resale price
import numpy as np
#floor_area_sqm, lease_commence_date, year, town, flat_type, storey_range, flat_model
X_user = np.array([[45.0,1978,2022,'ANG MO KIO','2 ROOM','10 TO 12','Improved']])
X_user_one_hot = one_hot.transform(X_user[:,[3,4,5,6]])
X_user1 = np.concatenate((X_user[:,[0,1,2]],X_user_one_hot),axis=1)
X_user2 = scale.transform(X_user1)
Resale_price = model_rand.predict(X_user2)
print(f"Resale price: {Resale_price[0]:.2f} SGD")
```
